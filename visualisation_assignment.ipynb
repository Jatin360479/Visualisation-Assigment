### Question 2.  Generate a dataset of random numbers. Visualize the distribution of a numerical values.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate a dataset of random numbers
data = np.random.randn(100)  # 100 random numbers from a standard normal distribution

# Create a DataFrame for easier plotting with Seaborn
df = pd.DataFrame({'value': data})

# Visualize the distribution using a histogram
sns.histplot(data=df, x='value', kde=True)  # kde=True adds a kernel density estimate curve
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

Question 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Generate a synthetic dataset with correlated features
data = pd.DataFrame({
    'Feature1': np.random.randn(100),
    'Feature2': 0.7 * np.random.randn(100) + 0.3 * np.random.randn(100),
    'Feature3': -0.5 * np.random.randn(100) + 0.5 * np.random.randn(100)
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Question 1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="Sine", line_width=2)

show(p)
```

### Question 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
### fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
### counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Change 'plot_height' to 'height'
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

### Question 4.  Create a Bokeh histogram to visualize the distribution of the given data.
### data_hist = np.random.randn(1000)
### hist, edges = np.histogram(data_hist, bins=30)


```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

show(p)
```
